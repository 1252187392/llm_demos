```python
#enconding: utf-8

from pdf_reader import PDFHelper
from retrieval import Retriever
```

```python
#if __name__ == '__main__':
pdf_reader = PDFHelper(max_pages=9000, content_max_len=10000000000)
pdf_path = "斗破苍穹.pdf"
pdf_reader.add_pdf(pdf_path, only_zh=True)
pdf_reader.init_chunks(sentence_num=12, only_zh=True)
print(pdf_reader.chunks[0])
```

```python
retriever = Retriever(max_length=1024)
embs = retriever.get_embs(pdf_reader.chunks, 128)
print(embs.shape)
```

```python
def get_topk(query, embs, k=10):
    top_k, scores = retriever.get_index(query, embs, k=k)
    print('topk:', top_k[:3], 'scores:', scores[:3])
    contexts = []
    for i in top_k:
        #print(f'chunk {i}: {pdf_reader.chunks[i]}')
        idx = pdf_reader.chunks_idx[i]
        context = ' '.join(pdf_reader.sentences[max(0, idx-6): idx+10])
        #print(f'context {context}')
        contexts.append(context)
    return contexts
```

```python
query = "萧炎终于晋级斗宗"
contexts = get_topk(query, embs, 64)
for content in contexts[:5]:
    print("---")
    print(content)
```

```python
from FlagEmbedding import FlagReranker
reranker = FlagReranker('BAAI/bge-reranker-v2-m3', use_fp16=True) # Setting use_fp16 to True speeds up computation with a slight performance degradation


# You can map the scores into 0-1 by set "normalize=True", which will apply sigmoid function to the score
score = reranker.compute_score(['query', 'passage'], normalize=True)
print(score) # 0.003497010252573502

# You can map the scores into 0-1 by set "normalize=True", which will apply sigmoid function to the score
scores = reranker.compute_score([['what is panda?', 'hi'], ['what is panda?', 'The giant panda (Ailuropoda melanoleuca), sometimes called a panda bear or simply panda, is a bear species endemic to China.']], normalize=True)
print(scores) # [0.00027803096387751553, 0.9948403768236574]
```

```python
def rerank(query, contexts, k=3):
    rerank_pairs = [[query, context] for context in contexts]
    scores = reranker.compute_score(rerank_pairs, normalize=True)
    #print(type(scores))
    #print(scores)
    idxs = scores.argsort()[::-1][:k]
    res = []
    for idx in idxs:
        res.append(contexts[idx])
    return res
```

```python
def get_prompt(query, recall_num=96, k=5):
    contexts = get_topk(query, embs, recall_num)
    contexts = rerank(query, contexts, k=k)
    prompt = '基于以下参考内容，回答用户问题。\n'
    for i in range(len(contexts)):
        prompt += f'参考内容{i}：{contexts[i]}\n'
        prompt += '\n'
    return prompt
```
